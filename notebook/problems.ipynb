## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import numpy as np
import random

def dice_experiment(num_time):
    count_events = 0

    for x in range(num_time):
      
        dice1 = random.randint(1, 6)
        dice2 = random.randint(1, 6)

        total = dice1 + dice2

     
        if total % 2 == 0 or total > 7:
            count_events += 1

        probability = count_events / num_time

    return probability

result= dice_experiment (1000) * 100

print(f"Probability_even_or_greater_than_7 is: {result} %")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
import random

def box_experiment(num_time=1000 ):
    count_white_red= 0
    count_all_same = 0

    for x in range(num_time):
        colors_balls =  []
        for x in range(5):
           colors_balls.append(ball_box[np.random.randint(0, 59)]) 

           if colors_balls.count("White") == 3 and colors_balls.count("Red") == 2:
              count_white_red += 1
    
           if colors_balls.count("White") == 5 or colors_balls.count("Red") == 5 or colors_balls.count("Green")== 5:
               count_all_same += 1

          
    probability_color = count_white_red / num_time
    probability_all = count_all_same / num_time
    
    return probability_color,  probability_all

result= box_experiment (1000) 


print(f"The probability of 3 white and 2 red is: {np.round(result[0] * 100, 2)}%")
print(f"The probability of all the same color is: {np.round(result[1] * 100, 2)}%")
```
